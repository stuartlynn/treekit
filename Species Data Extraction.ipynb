```python
from bs4 import BeautifulSoup
import requests
```

```python
soup = BeautifulSoup(requests.get('http://leafsnap.com/species/').text)
soup
#, attrs={'class','speciesTable'},
#            converters = {'Leaf': lambda x: print(type(x))}
#            )[0]
```

```python
table = soup.select_one('table.speciesTable')
```

```python
" dfsfd ".strip()
```

```python
def parse_row(row):
    entries = row.select('td')
    return {
        'leaf' : entries[0].select_one('img').attrs["src"],
        'flower':entries[1].select_one('img').attrs['src'] ,
        'fruit': entries[2].select_one('img').attrs['src'] ,
        'CommonName': entries[3].contents[0].strip(),
        'ScientificName': entries[4].select_one('a').contents[0].strip()
    }
headers  = [ row.contents  for row in table.select_one('thead').select("th")]

data  = [ parse_row(row)  for row in table.select_one('tbody').select("tr")]

import json
with open('species.json','w') as f:
    json.dump(data,f)
```
